```python
import pandas as pd
```

```python
#converting from
#int -> float is completely normal
#float -> int chops of decimal values and that can be considered data loss
#numeric -> object is dangerous as it can slow down performance. When the data type is numeric Pandas works fast using numPy operations, but when changed to object pandas falls back to pythonic loops.
#When pandas series includes multiple data types, it does not consider it in Numeric dtype it changes it to an object type
```

```python
#checking data type of series
s = pd.Series([1,2,3,4] , index = [10,20,30,40])
print(s.dtype)
```

```python
#changing data type of series
s.astype(float)
```

```python
#Upscaling -> The process where int automatically upgrades to float to store NaN values is called upscaling
s1 = pd.Series([10,20,30,40])
print(s1.dtype)
s1[2] = None
s1
```

```python
#Int64 vs int64
#int64 -> is numPy integer datatype which can't store NaN values so it upscales to float
#Int64 -> is Pandas nullable integer datatype which can store NaN values (coverts None to <NA>)
s2 = pd.Series([10,20,None,40], dtype = "Int64")
print(s2)
```
