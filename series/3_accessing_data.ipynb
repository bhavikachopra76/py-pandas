```python
import pandas as pd
```

```python
s = pd.Series([1,2,3,4,5] , index = ['a' , 'b', 'c', 'd', 'e'])
s
```

```python
#accesing values
s.values
```

```python
#accessing indexes
s.index
```

```python
#accesing data position-based
print(s.iloc[2])
```

```python
#accesing data index/label - based
print(s.loc['e'])
```

```python
#negative indexing
#this can only be done with 'iloc' as it is position based but performing it with 'loc' give a key-error
s1 = pd.Series([1,2,3,4] , index = ['a' , 'b', 'c', 'd'])
print(s1.iloc[-1])
```

```python
#slicing
#iloc slicing
s2 = pd.Series([10,20,30,40,50,60,70] , index = ['a' , 'b', 'c', 'd' , 'e', 'f', 'g'])
print(s2.iloc[0:4]) #included:excluded
```

```python
#loc slicing
print(s2.loc['d':'g']) #included:included
```

```python
#scalar access
#used when we need a single value only (optimized for scalar values)
print(s2.iat[2])
print(s2.at['c'])
```

```python
#if index is unique
prices = pd.Series([22150.25, 22150.30, 22150.50], index = ['09:15:00.001' , '09:15:00.001' , '09:15:00.002'])
print(prices)
print(prices.index.is_unique) #index should be unique for important trading calculations in future
```

```python
#sorted/unsorted labels
s3 = pd.Series([10,20,30,40], index=[103,100,102,101])
print(s3.loc[103:101]) #If the indexes are not sorted, slicing follows the order in which labels appear in the index
#103 (pos 0)
#100 (pos 1)
#102 (pos 2)
#101 (pos 3)
```

```python
print(s3.loc[102:103]) #if position 'a' > 'b' it returns an empty series
```

```python
s4 = pd.Series([10,20,30,40], index=[100,101,102,103])
print(s4.loc[102:100]) #even in a sorted array, the position of 'a' > 'b' it returns an empty series
```
