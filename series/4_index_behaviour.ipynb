```python
import pandas as pd
```

```python
#custom index
s = pd.Series([1,2,3,4,5] , index = ['a','b','c','d','e'])
s
```

```python
#index vs position
#index -> labels given to the data in the series
s.index
#position -> the position where data is located (starting from 0)
```

```python
# Replacing entire index -> We can not replace a single index in series
s.index = ['A' , 'B' , 'C' , 'D' , 'E']
s
```

```python
# Resetting index
# It is used to reset index to default integer-based indexing (0,1,2...) and convert the original index to a new column
# But, in series we can't reset index inplace as it will create a dataframe
# Pandas does not allow changing of datatype from series to dataframe
```

```python
prices = pd.Series([100,50,80,200,150] , index = ['Mon' , 'Tues' , 'Wed' , 'Thurs' , 'Fri'])
prices
```

```python
#giving name to series 
prices.name = "Prices"
```

```python
prices.reset_index() #temporarily resets the index
```

```python
#RangeIndex -> It is default integer index pandas create when you don't provide one
#0,1,2,3,4....
s1 = pd.Series([10,20,30])
print(type(s1.index)) #it stores value of index as [start stop step] as it would occupy a lot of memory to store indexes if there are elements from 1 to 999999 (so, it just stores [1 999999 1])
```

```python
#DateTime Index
dates = pd.date_range('16-02-2026' , periods = 5)
s2 = pd.Series([10,20,30,40,50] , index = dates) #now we can access the data by day
s2
```

```python
print(type(s2.index))
```

```python
#Reindex -> Forces data to match a new index structure
#Reordering data
s3 = pd.Series([10,20,30] , index = ['Infy' , 'TCS' , 'Reliance'])
print(s3)
s3 = s3.reindex(['Infy' , 'Reliance', 'TCS'])
print(s3)
```

```python
#Adding missing values
s4 = pd.Series([10,20,40] , index = [15,16,18])
s4 = s4.reindex([15,16,17,18])
s4
```

```python
#Changing index
s5 = pd.Series([10,20,30], index = ['A' , 'B' , 'C'])
s5 = s5.reindex(['A' , 'D' , 'C'])
s5
```

```python
#sorting index and values
s6 = pd.Series([1,3,4,2,1] , index = [10,80,60,20,40])
s6
```

```python
#sort index
s6.sort_index()
```

```python
#sort values
s6.sort_values()
```

```python
#same index returns a series not a scalar value
s7 = pd.Series([1,2,3,4] , index = ['A' , 'A' , 'B' , 'C'])
print(s7.loc['A'])
```
