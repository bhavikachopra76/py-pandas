```python
import pandas as pd
```

```python
s = pd.Series([1,2,3,4,5], index = ['a','b','c','d','e'])
s
```

```python
#conditional filtering
s[s>2]
```

```python
#multiple conditions
s1 = pd.Series([5, 15, 25, 55])
print(s1[(s1 > 10) & (s1 < 50)])
print(s1[(s1 > 10) | (s1 < 50)])
```

```python
#filtering vs where
s3 = pd.Series([1,2,3,4,5], index = ['a','b','c','d','e'])
print(s3[s3>2]) #removes the rows which don't match the condition
print(s3.where(s3>2)) #keeps structure, replaces values to NaN if they don't match the conditions
```

```python
#mask vs where
print(s3.mask(s3>2)) #replaces the values which satisfy the condition with NaN
print(s3.where(s3>2)) #replaces the values which don't satisfy the condition with NaN
```

```python
print(s3.mask(s3 > 2, 999))
#we can fill the NaN values with any other value 
```

```python
#between
print(s3.between(2,5)) #(both included)
```

```python
#isin -> checks membership
print(s3.isin([1,5,3]))
```
