```python
import numpy as np
import pandas as pd
```

```python
#None -> NaN
#It is considered a floating value in pandas and numpy
s = pd.Series([1,2,None,3,4,None])
s
```

```python
#checking for NaN values
s.isna()
```

```python
#filling NaN values
s.fillna(0)
```

```python
#aggregation ignoring NaN
print(s.sum()) #aggregation functions ignore NaN values and give the result because skipna = True by default
print(s.sum(skipna=False)) #if we do skipna = False it returns NaN (because anything + NaN = NaN)
```

```python
#Arithmetic keepin NaN
print(s+5) #arithemtic operations include NaN values and returns NaN
```

```python
#dropping NaN values
s.dropna()
```

```python
s1 = pd.Series([1,2,None,3,4,None])
s1.ffill() #forward fill
```

```python
s1.bfill() #backward fill
```

```python
#interpolation
s2 = pd.Series([1,2,None,4,None])
s2.interpolate() #fills the value by guessing through the surrounding values
```

```python
#detecting infinite values
s3 = pd.Series([1,2,np.inf, 3, -np.inf, 4])
np.isinf(s3)
```

```python
#replacing inf values
#NaN is considered better than infinity values as they are manageable but one inf value can ruin everything
#replacing with NaN
s3.replace([np.inf, -np.inf], np.nan)
```

```python
#replacing with other values
s3.replace([np.inf, -np.inf], 3)
```
