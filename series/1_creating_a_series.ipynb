```python
import pandas as pd
import numpy as np
```

```python
#creating a series from list
lst = [1,2,3,4,5,6]
s = pd.Series(lst)
s
```

```python
#creating a series from dictionary
d = {
    "Name" : "Bhavika",
    "Age" : 21,
    "Marks" : 98
}
s2 = pd.Series(d)
s2
```

```python
#creating a series with custom index
prices = pd.Series([100,50,80,200,150] , index = ['Mon' , 'Tues' , 'Wed' , 'Thurs' , 'Fri'])
prices
```

```python
#creating a series from numpy array
arr = np.array([10,20,30,40])
s3 = pd.Series(arr)
arr[2] = 60
print(arr)
s3
#Elements in numpy array and pandas series used to refer to the same memory location. But in newer pandas version by default 'copy = True' is already implemented while forming a series from a numpy array.
```

```python
#understanding numpy array
#If we do 'copy = False' the elements in both numpy array and Pandas series created from it would point to the same memory locations
arr2 = np.array([10,20,30])
s4 = pd.Series(arr2 , copy = False)
arr2[2] = 60
print(arr2)
print(s4)
s4.values.base is arr2
```

```python
#creating series with scalar value
s5 = pd.Series(5 , index = [10,20,30,40])
s5
```
