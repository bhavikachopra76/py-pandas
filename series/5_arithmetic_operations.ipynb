```python
import pandas as pd
```

```python
s = pd.Series([10,20,30,40,50])
s
```

```python
#Element wise multiplication
print(s*2)
```

```python
#Addition of 2 series
#Alignment by index
#Operations are performed by aligning elements to their indexes
#Element with index 'a' in 's1' will be added with Element with index 'a' in 's2'
s1 = pd.Series([1,2,3] , index = ['a' , 'b' , 'c'])
s2 = pd.Series([4,5,6] , index = ['a' , 'd' , 'c'])
print(s1 + s2)
```

```python
s3 = pd.Series([1,2,3] , index = ['a' , 'b' , 'c'])
s4 = pd.Series([4,5,6] , index = ['a' , 'b' , 'c'])
print(s3+s4)
```

```python
#add() mul() sub()
#They are different from normal arithmetic operators as we get fill_value with this 
print(s1.add(s2, fill_value = 0))
print(s1.mul(s2, fill_value = 2))
print(s2.sub(s1, fill_value = 1))
```

```python
print(s1.mean())
print(s1.std())
```

```python
#round values
s5 = pd.Series([1.2,2.3,3.4], index = [10,20,30])
s5.round()
```
