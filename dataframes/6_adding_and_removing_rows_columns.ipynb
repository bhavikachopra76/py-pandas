```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Name': ['Bhavika', 'Chaitanya', 'Dharya'],
    'Country': ['India', 'United States', 'India'],
    'Salary': [50000, 80000, 60000],
    'Age' : [21,22,20]
})

df
```

```python
#Add columns using scalar
df['Tax'] = 0.2
df
```

```python
#Column based on existing column
df['Bonus'] = df['Salary'] * 0.1
df
```

```python
#Column using condition
df['High Earner'] = df['Salary'] > 60000
df
```

```python
#Inserting column at any position
df.insert(0, 'EmployeeID' , [101,102,103])
df
```

```python
#Dropping columns
df.drop(columns = ['High Earner'] , inplace=True)
df
```

```python
#Adding rows
#Adding a single row
df.loc[len(df)] = [104 , 'Aarav' , 'Canada' , 75000 , 24 , 0.2 , 7500]
df
```

```python
#Adding multiple rows
new_rows = pd.DataFrame({
    'EmployeeID' : [105,106],
    'Name': ['Sandy' , 'Diva'],
    'Country': ['India', 'United States'],
    'Salary': [65000, 55000],
    'Age' : [24, 21],
    'Tax' : [0.2 , 0.2],
    'Bonus' : [6500, 5500]
})
df = pd.concat([new_rows , df] , ignore_index=True)
df
```

```python
df['EmployeeID'] = [101,102,103,104,105,106]
df
```

```python
#Dropping rows
df.drop(index = [0,1]) #requires inplace = true
```

```python
#drop by position (if index is not 0,1,2)
df.drop(df.index[5])  #requires inplace = true
```

```python
df.drop(columns = 'EmployeeID' , inplace = True)
df
```

```python
#dropping using conditions
filt = df['Salary'] < 65000
df.drop(df.index[filt])
```

```python
#Reindex
#reorder index
df = df.reindex([2,3,4,0,1,5])
df
```

```python
#reorder columns 
df = df.reindex(columns = ['Name' , 'Age' , 'Salary' , 'Bonus' , 'Tax' , 'Country'])
df
```
