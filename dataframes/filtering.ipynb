```python
import pandas as pd
```

```python
data = {
    "first_name" : ["Bhavika" , "Chaitanya" , "Dharya"],
    "last_name" : ["Chopra" , "Garg" , "Chopra"],
    "age" : [21,22,16],
    "marks" : [85,92,98]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
#creating mask
filt = df['last_name'] == 'Chopra'
#Applying mask
df[filt]
```

```python
#Only selected rows with single column
filt = df['marks'] > 90
df.loc[filt , 'marks']
```

```python
#Only selected rows with multiple columns
filt = df['marks'] > 90
df.loc[filt , ['first_name','marks']]
```

```python
#multiple conditions
#& -> Both conditions must be True
filt = (df['last_name'] == 'Chopra') & (df['marks'] > 90)
df.loc[filt , ['first_name','last_name','marks']]
```

```python
# | -> At least one condition must be True
filt = (df['last_name'] == 'Chopra') | (df['marks'] > 90)
df.loc[filt , ['first_name','last_name','marks']]
```

```python
# ~ -> Reverses boolean values
filt = df['last_name'] == 'Chopra'
df[~filt]
```

```python
# where -> keeps structure and gives NaN or any values to the False rows
df.where(df['age'] > 18 , 0)
```

```python
#mask -> keeps structure and gives NaN or any value to the True rows
df.mask(df['age'] > 18 , 0)
```

```python
#between
df[df['age'].between(18,50)]
```

```python
#isin
df[df['age'].isin([18,21,16])]
```
