```python
import pandas as pd
```

```python
trades = pd.DataFrame({
    "trade_id": [1, 2, 3 , 4],
    "product": ["CL", "NG", "ES" , "GC"],
    "quantity": [10, 5, 7 , 12]
})

products = pd.DataFrame({
    "product": ["CL", "NG", "ES"],
    "exchange": ["NYMEX", "NYMEX", "CME"]
})
```

```python
trades
```

```python
products
```

```python
#pd.merge() Basics
#inner
pd.merge(trades, products , on='product' , how= 'inner')
#Inner join keeps: Only rows that match in BOTH tables.
```

```python
#left
#Keep everything from left table.
pd.merge(trades,products , on='product' , how= 'left')
```

```python
#right
# Keep everything from right table.
pd.merge(trades,products , on='product' , how= 'right')
```

```python
#outer
#Keep EVERYTHING from both.
pd.merge(trades,products , on='product' , how= 'outer')
```

```python
#Mergin on index (matching indexes)
price_data = pd.DataFrame({
    "price": [100, 101, 102]
}, index=pd.to_datetime([
    "2024-01-01 09:00",
    "2024-01-01 09:01",
    "2024-01-01 09:02"
]))
sentiment_data = pd.DataFrame({
    "sentiment_score": [0.5, -0.2, 0.8]
}, index=pd.to_datetime([
    "2024-01-01 09:00",
    "2024-01-01 09:01",
    "2024-01-01 09:02"
]))
```

```python
price_data
```

```python
sentiment_data
```

```python
#Joining
price_data.join(sentiment_data)
```

```python
#concatenation
day1 = pd.DataFrame({
    "product": ["CL", "NG"],
    "price": [100, 200]
})

day2 = pd.DataFrame({
    "product": ["ES", "GC"],
    "price": [300, 400]
})
```

```python
day1
```

```python
day2
```

```python
pd.concat([day1, day2])
```
