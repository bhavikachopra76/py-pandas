```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Name': ['Bhavika', 'Chaitanya', 'Dharya'],
    'Country': ['India', 'United States', 'India'],
    'Salary': [50000, 80000, 60000],
    'Age' : [21,22,20]
})

df
```

```python
#apply() -> Series level
df['Name'].apply(len) #(Name is a series here)
#output is also a series
```

```python
#Custom transformation
df['Salary'].apply(lambda x : x + 1)
```

```python
#apply() -> Dataframe level
df.apply(len) #default axis = 0 (Rows collapse)
```

```python
df.apply(len , axis=1) #axis = 1 (Columns collapse)
```

```python
#map
#For series
df['Country'] = df['Country'].map({
    'India' : 'IN',
    'United States' : 'US',
})
df #But unmapped becomes NaN so there is a better alternative
```

```python
#Replace()
df['Country'] = df['Country'].replace({
    'IN' : 'India',
})
df #unmapped data remains the same
```
