```python
import pandas as pd
```

```python
data = {
    "first_name" : ["Bhavika" , "Chaitanya" , "Dharya"],
    "last_name" : ["Chopra" , "Garg" , "Chopra"],
    "age" : [21,22,16],
    "marks" : [85,92,98],
    "email" : ["BhavikaChopra@gmail.com" , "ChaitanyaGarg@gmail.com" , "DharyaChopra@gmail.com"]
}
df = pd.DataFrame(data)
```

```python
df.columns
```

```python
#Updating columns
#Renaming column names -> Method 1 (this can't be used if we have lots of columns)
df.columns = ['first name' , 'last name' , 'age' , 'marks' , 'email']
df
```

```python
#Renaming column names -> Method 2 (Renaming only the columns we want)
df.rename(columns={'marks' : 'Percentage'} , inplace = True)
df
```

```python
#Using string operations on columns
df.columns = df.columns.str.capitalize()
df.columns = df.columns.str.replace(' ' , '_')
```

```python
df
```

```python
#Updating Rows
#Updating Entire Row
df.loc[0] = ['Bhavika' , 'Chopra' , 21 , 90 , 'BhavikaChopra@gmail.com'] #It is un-necessary as we only changed one value
df
```

```python
#Updating specific cells
df.loc[2 , ['First_name' , 'Age']] = ['Anjali' , 45]
df
```

```python
#Updating a single cell (can use loc but at is optimized for accessing scalar values)
df.at[2, 'First_name'] = "Dharya"
df
```

```python
#Updating Entire Column
df['Email'] = df['Email'].str.lower()
df
```

```python
#Update using conditions
df.loc[df['Age'] > 20 , 'Percentage'] = 95
df
```

```python
#Update using multiple conditions
filt = (df['First_name'] == 'Dharya') & (df['Last_name'] == 'Chopra')
df.loc[filt , 'Age'] = 16
df
```

```python
#Changing index label
df.rename(index = {0:'A' , 1:'B' , 2:'C' }, inplace = True)
df
```

```python
#Broadcasting
#Scalar Broadcasting
df['Percentage'] = 0
df
```

```python
#Conditional Broadcasting
df.loc[df['First_name'] == 'Bhavika', 'Percentage'] = 92
df
```

```python
#Multiple Row Broadcasting
df.loc[['B', 'C'] , 'Percentage'] = [95,98]
df
```

```python
#Column-wise Broadcasting
df['Percentage'] = df['Percentage'] + 1
df
```
