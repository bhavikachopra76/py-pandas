```python
import pandas as pd

df = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance'],
    'Employee': ['Bhavika', 'Chaitanya', 'Dharya', 'Anjali', 'Dheeraj'],
    'Salary': [60000, 80000, 50000, 52000, 70000]
})

df
```

```python
#Groupby Basics
#Split
df.groupby('Department') #Returns a groupby object
```

```python
#Apply
df.groupby('Department')['Salary'].mean()
```

```python
#Basic Aggregation
df.groupby('Department')['Salary'].median()
```

```python
df.groupby('Department')['Salary'].count() #counts non-none values
```

```python
df.groupby('Department')['Salary'].size() #includes none values
```

```python
df.groupby('Department')['Salary'].min()
```

```python
df.groupby('Department')['Salary'].max()
```

```python
#Multiple Aggregations on One Column
df.groupby('Department')['Salary'].agg(['mean' , 'median' , 'min' , 'max'])
```

```python
#Different Aggregations for Different Columns
df.groupby('Department').agg({
    'Salary' : ['mean'],
    'Employee' : ['count']
})
```

```python
#Custom Column Names
df.groupby('Department').agg(
    avg_salary = ('Salary' , 'mean'),
    max_salary = ('Salary' , 'max'),
    total_employees = ('Employee' , 'count')
)
```

```python
#Grouping my multiple columns
df2 = pd.DataFrame({
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance'],
    'Employee': ['Bhavika', 'Chaitanya', 'Dharya', 'Anjali', 'Dheeraj'],
    'Gender' : ['F' , 'M' , 'M' , 'F' , 'M'],
    'Salary': [60000, 80000, 50000, 52000, 70000]
})
df2
```

```python
grouped = df2.groupby(['Department' , 'Gender'])['Salary'].mean()
grouped
```

```python
grouped.index #This has a multiIndex
```

```python
#as_index = True/False
df.groupby('Department' , as_index = False)['Salary'].max()
```
