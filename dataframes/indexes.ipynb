```python
import pandas as pd
```

```python
#An Index is the row label system of a DataFrame.
#Columns label the vertical axis
# Index labels the horizontal axis
#When we create a DataFrame without specifying index: It assigns a rangeIndex (start = 0, stop = n, step = 1 )
```

```python
data = {
    "first_name" : ["Bhavika" , "Chaitanya" , "Dharya"],
    "last_name" : ["Chopra" , "Garg" , "Chopra"],
    "age" : [21,22,16]
}
df = pd.DataFrame(data)
```

```python
df
```

```python
#inspecting index
df.index
```

```python
#length of index
len(df.index)
```

```python
#check if index is unique
df.index.is_unique
```

```python
df2 = pd.DataFrame({
    "price" : [100,200,300]
} , index = ['a','b','c'])

dates = pd.date_range('16-02-2026' , periods = 3)
df3 = pd.DataFrame({
    "price" : [100,200,300]
} , index = dates)
```

```python
#types of indexes
# 1st -> RangeIndex (assigned by default if we don't mention)
print(type(df.index))
# 2nd -> Index (contains int, float, str)
print(type(df2.index))
# 3rd -> DateTime index (gives date)
print(type(df3.index))
```
