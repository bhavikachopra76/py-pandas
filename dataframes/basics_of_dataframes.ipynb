```python
import pandas as pd
```

```python
#creating dataframes from lists
#all the list should be of same size else it will throw an error
data = {
    "first_name" : ["Bhavika" , "Chaitanya" , "Dharya"],
    "last_name" : ["Chopra" , "Garg" , "Chopra"],
    "age" : [21,22,16]
}
df = pd.DataFrame(data)
df
```

```python
#creating dataframes from series
#unequal length works for series because they have labels, and they can be used for alignment
#But Lists don't have index labels for alignment so, for lists pandas assumes row 0 of 'A' matches row 0 of 'B'
price = pd.Series([10,50,100,150])
qnt = pd.Series([15,5,2,1])
df2 = pd.DataFrame({
    "prices" : price,
    "quantity" : qnt
})
df2
```
