```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'Name': ['Bhavika', 'Chaitanya', 'Dharya'],
    'Country': ['India', 'United States', 'India'],
    'Salary': [50000, 80000, 80000],
    'Age' : [21,22,20]
})

df
```

```python
#Sorting by column values
df.sort_values(by = 'Salary')
```

```python
#Descending
df.sort_values(by = 'Salary' , ascending = False)
```

```python
#Sort by multiple columns
df.sort_values(['Salary' , 'Age'])
```

```python
#Different order per column
df.sort_values(['Salary' , 'Age'] , ascending = [False,True] , inplace=True)
```

```python
#Sort by index
df.sort_index(inplace = True)
df
```

```python
df.sort_index(axis = 1)
```

```python
df.nlargest(2,'Salary')
```

```python
df.nsmallest(2,'Salary')
```
