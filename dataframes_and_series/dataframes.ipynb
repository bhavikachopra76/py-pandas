```python
import pandas as pd
```

```python
#Creating Dictionary from List (all the list should be of same size else it will throw an error )
data = {
    "Name" : ["Bhavika", "Chaitanya" , "Dharya"],
    "Age" :[21,22,16],
    "Marks" :[85, 92, 98]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#Creating dataframe using series
s1 = pd.Series([1,2,3,4])
s2 = pd.Series([5,6]) #here series can be of different sizes
df2 = pd.DataFrame({
    "A" : s1,
    "B" : s2
})
print(df2)
#unequal length works for series because they have labels, and they can be used for alignment
#But Lists don't have index labels for alignment so, for lists pandas assumes row 0 of 'A' matches row 0 of 'B'
```
