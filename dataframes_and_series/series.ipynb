```python
import pandas as pd
```

```python
#Creating and printing a series
s = pd.Series([10,20,30])
s
```

```python
#elements in series
s.values
```

```python
#indexes of elements
s.index
```

```python
#Index here is different from position, indexes can be changed here.
s1 = pd.Series([2,4,6,8], index=[10,20,30,40])
s1
```

```python
#how to access elements from series
#There are 2 options
print(s1.iloc[0]) #accessing by position
```

```python
print(s1.loc[40]) #accessing by index
```

```python
#Addition of 2 series
#Pandas does not add by position, it adds by Index Label
s2 = pd.Series([1,2,3], index = ['a','b','c'])
s3 = pd.Series([4,5,6], index = ['a','d','c'])
print(s2)
print(s3)
```

```python
print(s2+s3)
```

```python
#creating series from dictionary
data = {'A': 100, 'B': 200, 'C': 300}
se1 = pd.Series(data)
print(se1)
```

```python
#In trading terms
prices = {
    "NIFTY" : 22100,
    "BANKNIFTY" : 46800,
    "FINNIFTY" : 21000
}
se2 = pd.Series(prices)
print(se2)
```

```python
#Handling missing values
se3 = pd.Series([1,2,None,3,None,4,5])
print(se3) #None is directly converted to NaN and data type of NaN is float in numpy/pandas
```

```python
#Finding NaN values
se3.isna()
```

```python
#Changing values of NaN
se3.fillna(0)
```

```python
#Mathematical Operations
print(se3 * 4) #Arithmetic operations don't skip NaN values
```

```python
#Aggregation
print(se3.sum()) #skips the NaN values as by default skipna = True in Pandas
```

```python
#Boolean filtering
se4 = pd.Series([10,20,30,40,50,60])
se4 > 20
```

```python
se4[se4>20]
```

```python
#Index Properties
#we can't change a single index, but we can replace entire index
se5 = pd.Series([10,20,30])
print(se5)
```

```python
se5.index = ['A' , 'B' , 'C']
print(se5)
```

```python
 #Reset Index
price = pd.Series([100,150,200,250,180] , index=["Mon" , "Tue" , "Wed" , "Thur" , "Fri"])
price.name = 'Prices'
price
```

```python
price.reset_index()
```

```python
price #price.reset_index() temporarily changes the index it does not affect the original data frame
```

```python
price.reset_index(inplace = True) #this would give an error because inplace+true would change the series into a dataframe
```
