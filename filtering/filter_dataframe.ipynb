```python
import pandas as pd
```

```python
data = {
    "First name": ["Bhavika", "Chaitanya", "Dharya"],
    "Last name": ["Chopra", "Garg", "Chopra"],
    "Email": ["bhavikachopra@gmail.com", "chaitanyagarg@gmail.com", "dharyachopra@gmail.com"]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
#filtering using loc
filt = df['Last name'] == 'Chopra'
```

```python
df.loc[filt]
```

```python
#accesing columns using the same filter
df.loc[filt , 'Email']
```

```python
#using & | ~
filt2 = (df['First name'] == 'Bhavika') & (df['Last name'] == 'Chopra')
```

```python
df.loc[filt2]
```

```python
filt3 = (df['First name'] == 'Bhavika') | (df['Last name'] == 'Chopra')
```

```python
df.loc[filt3]
```

```python
df.loc[~filt3]
```
